```python
import pandas as pd
import os
import time

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier as classifier

from sklearn.metrics import f1_score
```

### 時間の計測スタート

```python
start = time.time()
```

### データの読込

```python
path = os.getcwd() + "/"
```

```python
train = pd.read_csv(path + "train.csv")
train.head(6)
```

```python
train.shape
```

```python
test = pd.read_csv(path + "test.csv")
test.head(6)
```

```python
test.shape
```

### 統計量の表示

```python
train.describe()
```

```python
test.describe()
```

### 欠損値の処理

```python
# 欠損値カウント
train.isnull().sum()
```

```python
# 欠損値カウント
test.isnull().sum()
```

```python
# trainデータのAGE が 欠損値の列を除く
train = train.dropna(subset=["Age"])
train.head(6)
```

```python
# testデータの Age の欠損値を平均値に置き換える
test["Age"] = test["Age"].fillna(test["Age"].mean())[12]
test.head(6)
```

```python
# testデータの Fare の欠損値を平均値に置き換える
test["Fare"] = test["Fare"].fillna(test["Fare"].mean())[12]
test.head(6)
```

### 相関係数の表示

```python
train.corr()["Survived"]
```

### 特徴量の選定

```python
x = train[["Pclass", "Age", "SibSp", "Parch", "Fare"]]
x.head(6)
```

```python
y = train["Survived"]
y.head(6)
```

```python
test_ft = test[["Pclass", "Age", "SibSp", "Parch", "Fare"]]
test_ft.head()
```

### データの分割

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

### 学習モデルの作成

```python
clf = classifier(n_estimators = 100, random_state=42)
```

```python
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
y_pred[0:6]
```

### 誤差

```python
f1_score(y_pred, y_test)
```

### 予測データの作成

```python
pred = clf.predict(test_ft)
pred[0:6]
```

```python
submission = pd.read_csv(path + "gender_submission.csv")
submission["Survived"] = pred
submission.head(6)
```

### csvの出力

```python
submission.to_csv('submission_rf.csv', index=False)
```

### 経過時間

```python
elapsed_time = time.time() - start
print(elapsed_time)
```
